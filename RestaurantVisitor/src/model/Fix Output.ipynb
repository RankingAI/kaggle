```python
import numpy as np
import pandas as pd
import os,sys,time

DataBaseDir = '../../data/MM2'
TreeNum = 200
kfold = 5

for fold in range(kfold):
    OutputDir = '../../data/MixedModels2/l1/kfold/%s' % fold
    if(os.path.exists(OutputDir) == False):
        os.makedirs(OutputDir)

start = time.time()
for t in range(TreeNum):
    TreeInputDir = '%s/l1/%s' % (DataBaseDir, t)
    for fold in range(kfold):
        InputDir = '%s/kfold/%s' % (TreeInputDir, fold)
        for mod in ['valid', 'holdout', 'test']:
            df = pd.read_csv('%s/%s_lgb_l2.csv' % (InputDir, mod))
            df = df.rename(index= str, columns= {'lgb_l2': 'lgb_l2_%s' % t})
            OutputFile = '../../data/MixedModels2/l1/kfold/%s/%s_lgb_l2_%s.csv' % (fold, mod, t)
            df.to_csv(OutputFile, index= False)
    end = time.time()
    print('tree %s done, time elapsed %ss' % (t, (end - start)))
```
