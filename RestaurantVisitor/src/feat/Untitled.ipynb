```python
import pandas as pd
import datetime

InputDir = '../../data/raw'
data = pd.read_csv('%s/date_info.csv' % InputDir, parse_dates=['calendar_date']).rename(columns={'calendar_date': 'visit_date'})

start = datetime.datetime(2016, 12, 25) - datetime.timedelta(days= 7)
end = datetime.datetime(2016, 12, 25)
data['before_christmas'] = data['visit_date'].apply(lambda dt: ((dt <= end) and (dt >= start)))
data['before_christmas'] = data['before_christmas'].astype(int)
start = datetime.datetime(2016, 12, 25)
end = datetime.datetime(2016, 12, 25) + datetime.timedelta(days= 7)
data['after_christmas'] = data['visit_date'].apply(lambda dt: ((dt <= end) and (dt >= start)))
data['after_christmas'] = data['after_christmas'].astype(int)
print(data['before_christmas'].value_counts())
```
